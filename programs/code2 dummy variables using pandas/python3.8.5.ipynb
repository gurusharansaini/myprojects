```python
import pandas as pd
```

```python
df=pd.read_csv("dummy_variable_data.csv")
df
```

```python
dummies=pd.get_dummies(df.Town)
dummies
```

```python
df_dummies=pd.concat([df,dummies],axis='columns')
df_dummies
```

```python
df_dummies.drop(['Town','Western windows'],axis='columns',inplace=True)
df_dummies
```

```python
x=df_dummies.drop('Price',axis='columns')
x
```

```python
y=df_dummies.Price
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[]])
```
